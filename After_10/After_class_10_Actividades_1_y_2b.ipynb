```python
# Importar las bibliotecas necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cargar el conjunto de datos "diamonds" de Seaborn
diamonds = sns.load_dataset("diamonds")
```

```python
# Visualizar las primeras filas del DataFrame
diamonds.head(3)
```

```python
# Veamos la estructura del dataset
diamonds.info()
```

```python
# Analizamos algunos datos estadísticos
diamonds["price"].describe()
```

```python
# Crear un boxplot que muestre cómo se distribuyen los valores de la variable numérica
# 'price' sobre la variable categórica 'color'
sns.boxplot(y='price', x ="color", data=diamonds)
# sns.boxplot(y='price', x= 'clarity', data=diamonds) # o x='cut'
plt.show()
```

```python
# Crear un gráfico de distribución de los precios con Seaborn
# (ver cómo agregar lineas verticales de referencias)
plt.figure()
sns.histplot(x="price", data=diamonds, kde=True, bins = 10)

# calcular la media y el desvio
media = diamonds["price"].mean()
mediana = diamonds["price"].median()
desvio = diamonds["price"].std()

# Agregar linear verticales
plt.axvline(media, color='red', linestyle='--', label=f'Media: {media:.2f}')
plt.axvline(mediana, color='green', linestyle='--', label=f'Mediana: {mediana:.2f}')

# agregar lineas verticales de mean +- 1 desvio
plt.axvline(media + desvio, color='blue', linestyle='--', label=f'Media + 1 desvio: {media + desvio:.2f}')
plt.axvline(media + 2*desvio, color='orange', linestyle='--', label=f'Media + 2 desvio: {media + 2*desvio:.2f}')

plt.legend()
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.title("Distribución de Precios de Diamantes")
plt.show()
```

```python
# Crear un gráfico de distribución por color
# puedo usar hist? o tendré que usar otro recurso?
sns.histplot(x="color", data=diamonds)
plt.show()
```

```python
# Crear un gráfico de distribución por color
# puedo usar hist? o tendré que usar otro recurso?
sns.countplot(x="color", data=diamonds)
plt.show()
```

```python
# Importar las bibliotecas necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cargar la base de datos "titanic"
titanic = sns.load_dataset("titanic")
```

```python
# Visualizar las primeras filas de la base de datos
titanic.head()
```

```python
# Realizar un análisis estadístico básico de la tasa de supervivencia por clase de pasajero
titanic_pp1 = titanic.copy()
titanic_pp1= titanic_pp1[["pclass", "survived"]].dropna()
titanic.groupby(['pclass'])['survived'].mean()
titanic['survived'].mean()
# titanic.groupby('sex')['survived'].mean()
```

```python
titanic_pp1.info()
```

```python
titanic["age"].describe()
```

```python
# Crear un histograma con la distribucion de las edades
sns.histplot(x="age", data=titanic)
plt.show()
```

```python
# Crear un gráfico de barras que muestre la tasa de supervivencia por clase
sns.barplot(x="pclass", y="survived", data=titanic)
plt.show()
```

```python
# Analizar si existe alguna correlación entre la edad de los pasajeros y su tasa de supervivencia
sns.boxplot(x="survived", y="age", data=titanic)
plt.show()
```

```python
# Bonus track 1: análisis de correlación con un mapa de calor
import seaborn as sns
import matplotlib.pyplot as plt

corr = titanic[['age', 'survived']].corr()

plt.figure(figsize=(5, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlación entre Edad y Supervivencia')
plt.show()
```

```python
# Bonus track 2: Análisis de regresión logística
plt.figure(figsize=(8, 6))
sns.regplot(x='age', y='survived', data=titanic, logistic=True, scatter_kws={'alpha':0.5})
plt.title('Probabilidad de supervivencia según la edad')
plt.xlabel('Edad')
plt.ylabel('Probabilidad de supervivencia')
plt.show()
```
